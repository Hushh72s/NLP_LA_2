```python
corpus = [
    "Stocks rally as market optimism grows",
    "New study shows health benefits of coffee",
    "Sports events cancelled due to heavy rain",
    "Tech companies innovate rapidly this year",
    "Economy slows as inflation concerns rise"
]
for i in corpus:
  print(i)
```

# **Text Representation using Traditional Methods**

### ***1.Bag Of Words***

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer()
bow=vectorizer.fit_transform(corpus)
print(f"Features Names: \n{vectorizer.get_feature_names_out()}")
print("")
print(f"TF-IDF Matrix Shape: \n{bow.shape}")
print("")
print(f"BoW Matrix: \n{bow.toarray()}")
```

### ***2.TF-IDF***

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
tfidf=vectorizer.fit_transform(corpus)
print(f"Feature names: \n{vectorizer.get_feature_names_out()}")
print("")
print(f"TF-IDF Matrix Shape: \n{tfidf.shape}")
print("")
print(f"TF-IDF Matrix: \n{tfidf.toarray()}")
```

### ***3. N-Grams***

```python
bigram_vec = CountVectorizer(ngram_range=(2, 2))
bigram_matrix = bigram_vec.fit_transform(corpus)

print("\n=== BI-GRAM REPRESENTATION ===")
print("Shape:", bigram_matrix.shape)
print("Sample features:", bigram_vec.get_feature_names_out())

trigram_vec = CountVectorizer(ngram_range=(3, 3))
trigram_matrix = trigram_vec.fit_transform(corpus)

print("\n=== TRI-GRAM REPRESENTATION ===")
print("Shape:", trigram_matrix.shape)
print("Sample features:", trigram_vec.get_feature_names_out())
```

## **Analysis of How Vectorization Affects the Feature Space**

### **1. Bag of Words (Count Vectorizer)**
- Produces one feature per unique token.  
- Feature space size grows with vocabulary size.  
- Uses simple frequency counts with no weighting.  
- Can be influenced by very common but uninformative words unless stop-words are removed.

---

### **2. TF-IDF Representation**
- Produces a similar number of features as Bag of Words (same vocabulary).  
- Assigns importance scores rather than raw counts.  
- Reduces the weight of terms that appear frequently across many documents.  
- Useful when emphasizing distinctive or meaningful terms.

---

### **3. N-Grams (Bi-grams / Tri-grams)**
- Increases the feature space significantly because token combinations become features.  
- Captures context by representing short phrases (e.g., *'as inflation' 'as market'* ).  
- Can produce very sparse, high-dimensional vectors.  
- Important to manage with caution, especially when working with small datasets or limited computational resources.

```python
# 1. Train a small Word2Vec model
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

# Simple training corpus
sentences = [
    ["king", "queen", "royal", "palace"],
    ["man", "woman", "person"],
    ["apple", "banana", "fruit", "food"],
    ["king", "man"],
    ["queen", "woman"],
    ["apple", "banana"],
]

# Train Word2Vec model
model = Word2Vec(sentences, vector_size=50, window=3, min_count=1)
wv = model.wv
```

```python
words = ["king", "queen", "man", "woman", "apple", "banana"]

for w in words:
    print(f"\nMost similar to '{w}':")
    print(wv.most_similar(w, topn=3))
```

```python
# Get vectors
vectors = np.array([wv[w] for w in words])

# PCA to 2D
pca = PCA(n_components=2)
coords = pca.fit_transform(vectors)

# Plot
plt.figure(figsize=(8,5))
plt.scatter(coords[:, 0], coords[:, 1])

for i, word in enumerate(words):
    plt.annotate(word, (coords[i,0], coords[i,1]), fontsize=12)

plt.title("Word2Vec Embeddings (PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

## Interpretation of the Word Embedding Visualization

The 2D plot shows clear semantic clustering among the selected words:

### **1. Human / Royalty Cluster**
- **king**
- **queen**
- **man**
- **woman**

These words appear near one another because the model learns that they share
common contexts. Pairs such as *king–queen* and *man–woman* represent
gender-based or role-based relationships. Their proximity reflects how the
Word2Vec algorithm groups concepts that frequently occur in similar linguistic
environments.

### **2. Fruit / Food Cluster**
- **apple**
- **banana**

These words are positioned together but separate from the human-related
cluster. The model associates them based on their appearance in similar
contexts involving food, eating, or fruit categories.

### **3. Semantic Structure**
The spacing between the clusters highlights the model’s ability to distinguish
between unrelated concept groups (people vs. fruits). At the same time,
it draws similar items closer together within each group, capturing both
category membership and contextual usage patterns.
